```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
data = {
    'Month': ['April', 'May', 'June', 'July', 'August', 'September'],
    'Sales': [7646775, 8449335, 8629201, 8666657, 8781439, 8235425]
}

df = pd.DataFrame(data)

month_map = {'April': 1, 'May': 2, 'June': 3, 'July': 4, 'August': 5, 'September': 6}
df['Month_num'] = df['Month'].map(month_map)

X = df[['Month_num']]
y = df['Sales']

model = LinearRegression()
model.fit(X, y)

october_prediction = model.predict([[7]])

months = ['April', 'May', 'June', 'July', 'August', 'September', 'October']
sales_data = df['Sales'].values.tolist() + [october_prediction[0]]

plt.figure(figsize=(10, 6))
plt.plot(months, sales_data, marker='o', linestyle='-', color='blue', label='Sales (Actual & Forecast)')
plt.axvline(x='September', color='gray', linestyle='--', alpha=0.6, label='Forecast Boundary')

plt.title("Food and Drinks Sales Forecast in Canada", fontsize=14)
plt.xlabel("Months", fontsize=12)
plt.ylabel("Sales (in units)", fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

print(f"Predicted sales for October: {october_prediction[0]}")
```
